```python
import requests
import time

def fetch_articles(api_key, cse_id, query, total_results=100):
    articles = []
    start_index = 1
    while len(articles) < total_results:
        url = f"https://www.googleapis.com/customsearch/v1?key={api_key}&cx={cse_id}&q={query}&start={start_index}"
        response = requests.get(url)
        results = response.json()
        items = results.get('items', [])
        for item in items:
            articles.append({
                'title': item.get('title'),
                'link': item.get('link')
            })
        if not items or len(articles) >= total_results:
            break
        start_index += 10
    return articles[:total_results]
```

```python
import requests
from bs4 import BeautifulSoup

# URL 목록이 저장된 파일 열기
input_file_path = 'articles.txt'  # 입력 파일 경로
output_file_path = 'article_scrap.txt'  # 출력 파일 경로

with open(input_file_path, 'r', encoding='utf-16') as file:
    urls = [line.strip() for line in file if line.startswith('http')]

# 추출된 텍스트를 저장할 파일 열기
with open(output_file_path, 'w', encoding='utf-8') as output_file:
    for url in urls:
        try:
            response = requests.get(url)
            soup = BeautifulSoup(response.content, 'html.parser')
            # 여기서는 단순화를 위해 전체 페이지 텍스트를 추출합니다.
            # 필요에 따라 soup.find() 또는 soup.find_all()을 사용하여
            # 특정 요소의 텍스트만 추출할 수 있습니다.
            text = soup.get_text(separator='\n', strip=True)
            # URL 및 추출된 텍스트를 출력 파일에 작성
            output_file.write(f"URL: {url}\nExtracted Text: {text[:1000]}\n\n---\n\n")
            print(f"Text extracted and saved from {url}")
        except Exception as e:
            print(f"Error fetching {url}: {e}")

print("All done!")
```

```python
import re

def remove_non_korean_from_file(input_file_path, output_file_path):
    with open(input_file_path, 'r', encoding='utf-8') as file:
        content = file.read()
    cleaned_content = re.sub(r'[^가-힣\s]', '', content)
    with open(output_file_path, 'w', encoding='utf-8') as file:
        file.write(cleaned_content)

# 파일 경로 설정
input_file_path = 'article_scrap.txt'  # 입력 파일 경로
output_file_path = 'article_cleaning.txt'  # 출력 파일 경로

# 함수 호출
remove_non_korean_from_file(input_file_path, output_file_path)

print("Processing completed. Check the output file.")
```

```python
file_path = 'article_cleaning.txt'  

with open(file_path, 'r', encoding='utf-8') as file:
    content = file.read()

content_clean = re.sub(r'\b[가-힣]\b', '', content)
tokens = content_clean.split()

print(f"Total number of tokens: {len(tokens)}")
```

```python
from collections import Counter

token_counts = Counter(tokens)
filtered_tokens = [token for token in tokens if token_counts[token] > 2]

print(f"Total number of tokens: {len(filtered_tokens)}")
```

```python
output_file_path = 'tokens.txt'
with open(output_file_path, 'w', encoding='utf-16') as file:
    for token in filtered_tokens:
        file.write(token + '\n')

print("Finish")
```
