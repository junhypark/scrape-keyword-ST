```python
# '교통 체증',
#     '주택 가격',
#     '범죄 및 안전 문제',
#     '소음 문제',
# 1st try - 저 4개만 온전히 다운로드
# fix: status code 거르기
# prob: 중간부터는 img src를 주질 않았음
# 이 문제점 해결해야 할듯
# proxy가 도움을 줄수도? = 켜져있었음.. ㅋ


    # '노숙자',
    # '소수자 및 저소득층 지역 사회 간의 격차',
    # '젠트리피케이션',
    # '커뮤니티 공간 부족',
# 2nd try - 저 4개만 온전히 다운로드
# 아마도 몇몇 이미지 다운로드하면 자동으로 block 시키는 듯?
# 그렇다면 status code로 체크하는것 또한 의미가 없음 ㅋ

pq = [
    '노인 우대',
    '환경 오염'
]
```

```python
final = []
count = 0
img_dest = "C:/Users/purpl/OneDrive/Desktop/Stan_2024/scrape-keyword-ST/image_scrape/images3/"
for words in pq:
    result = []
    j = 0
    for i in range(0, 20):
        resp = get('https://www.google.no/search?q='+words+'&client=opera&hs=cTQ&source=lnms&tbm=isch&sa=X&safe=active&ved=0ahUKEwig3LOx4PzKAhWGFywKHZyZAAgQ_AUIBygB&biw=1920&bih=982&start='+str(count))
        
        if resp.status_code != 200:
            print("status code down")
            time.sleep(100)
        
        dom = BeautifulSoup(resp.text, 'html.parser')
    
        for raw_img in dom.find_all('img'):
            link = raw_img.get('src')
            if link:
                try:
                    urllib.request.urlretrieve(link, img_dest+words+str(j)+'.jpg')
                    j += 1
                except:
                    print("error!", link)
                    pass

                result.append(link)

        count += 10
        time.sleep(10)
    final.append(result)
```

```python
import glob
import random
from PIL import Image
```

```python
input_keyword = input()
# keyword를 받게 되면 input_keyword로 저장

path = './images/'
rd_num = [random.randint(0, 399) for i in range(0,4)]
# 400개의 image들 중에서 4개 추출

image1 = Image.open(path+input_keyword+str(rd_num[0])+'.jpg')
image2 = Image.open(path+input_keyword+str(rd_num[1])+'.jpg')
image3 = Image.open(path+input_keyword+str(rd_num[2])+'.jpg')
image4 = Image.open(path+input_keyword+str(rd_num[3])+'.jpg')
# 해당 랜덤 이미지들을 선언

image_size = image1.size    # 모든 image들의 크기는 동일하므로 / size를 선언
new_image = Image.new('RGB', (2*image_size[0], 2*image_size[1]), (250, 250, 250))
# 빈 도화지를 만드는데 image_size에 2배씩 하여 만듦

new_image.paste(image1, (0,0))
new_image.paste(image2, (image1.size[0],0))
new_image.paste(image3, (0, image1.size[1]))
new_image.paste(image4, (image3.size[0], image1.size[1]))
# 좌표평면처럼 하나씩 붙여서 보여줌
# 1 | 2
#-------
# 3 | 4
# 와 같은 형식으로 보여짐

new_image.save('result.jpg', 'JPEG')
# 빈 도화지위에 넣어준 것을 저장
```
